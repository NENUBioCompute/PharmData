```python
document_count = drugs_mapping.collection.count_documents({})
document_count
```

```python

object_ids = []
for doc in drugs_mapping.collection.find():
    db_num = len(doc['dblink'])
    # print(len(doc['dblink']))
    if db_num>=3:
        object_ids.append(doc['_id'])
print(len(object_ids))
object_ids[0]
```

```python
from bson.objectid import ObjectId
new_col_name = 'drug'
new_collection = DBconnection(cfgfile,db_name,new_col_name)
new_collection.collection.delete_many({})
for object_id in object_ids:
    
    object_id = ObjectId(object_id)
    queried_doc = drugs_mapping.collection.find_one({'_id': object_id})
    
    new_collection.collection.insert_one(queried_doc)
    # print(queried_doc)
    
```
