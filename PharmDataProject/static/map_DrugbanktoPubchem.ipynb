```python
import pandas as pd
import gzip

# 指定文件路径
file_path = '/home/zhaojingtong/mapdata/src6src22.txt.gz'

# 使用gzip模块解压缩并读取文件
with gzip.open(file_path, 'rt') as f:
    # 假设文件是以制表符分隔的，可以根据实际情况修改delimiter参数
    df = pd.read_csv(f, delimiter='\t', header=None)

# 确保数据框至少有两列
if df.shape[1] < 2:
    raise ValueError("数据框至少需要两列")

# 将第一列作为键，第二列作为值转换为字典
data_dict = dict(zip(df.iloc[:, 0], df.iloc[:, 1]))

# 打印字典
print(data_dict)

```
